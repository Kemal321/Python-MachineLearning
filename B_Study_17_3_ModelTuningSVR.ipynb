```python
import yfinance as yf
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
import matplotlib.pyplot as plt 
from sklearn.model_selection import GridSearchCV
import warnings
```

```python
data = yf.download("THYAO.IS",start="2022-08-01",end="2022-09-01")
data2 = data.copy()
data2 = data2.reset_index()
print(data2)
```

```python
# Gün bilgisibi alalım
data2["Day"] = data2["Date"].astype(str).str.split("-").str[2]
data2.head()
```

```python
y = data2["Adj Close"] # kapanış fiyatlarını aldık
X = data2["Day"]

y = np.array(y).reshape(-1,1)
X = np.array(X).reshape(-1,1)

scalery = StandardScaler()
scalerX = StandardScaler()

y = scalery.fit_transform(y)
X = scalery.fit_transform(X)

model1 = SVR(kernel="rbf")
model1.fit(X,y)
```

Buraya kadar bir önceki çalışmadan yaptık zaten şimdi model tuning yani hiperparametrelerin optimize edilmesi konusuna bakacağız. Burada GridSearchCV denilen tekniği kullanacağız. Kontrol edeceğimiz çok fazla parametre yok. Düzeltme parametresi C, Gamma ve Kernek ile deneyeceğiz 

```python
# bu nokta da grid için search yapmak istediğimiz değerleri veya aralıkları vermemiz lazım 
# ben belirli rakamlar vereceğim
warnings.filterwarnings(action="ignore", category=DeprecationWarning) 
params = {"C":[1,10,100,1000,10000],"gamma":[1,0.1,0.01,0.001],"kernel":["rbf","poly","linear"]} 
# sözlük tipinde bu şekilde oluşturuyoruz. Aralık içinde dökümantasyondan öğrenin kolay zaten. 
tuning = GridSearchCV(estimator=SVR(),param_grid=params,cv=10)
tuning.fit(X,y)
print(tuning.best_params_)
# çıktı olarak bunu döndü yani verilerimize en uygun olanlar bunlar
# {'C': 100, 'gamma': 1, 'kernel': 'rbf'} olarak döndü
# em iyi olup olmadığına güvenmeyecek değilim :DD bu kadar 
```
