```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import sklearn.metrics as mt
```

```python
data = pd.read_excel(r"C:\Users\Kemalettin\Desktop\anlasekon\Python ile Makine Öğrenmesi\B_Data_3_temperature.xlsx")
data2 = data.copy()
data2.head()
```

```python
y = data2["Verim"]
X = data2["Sıcaklık"]

plt.scatter(X,y)
plt.show()
```

```python
print(data2.corr())
```

```python
# Linear olmadığına göre non linear denemeyelim
pol = PolynomialFeatures(degree=2) # burada degree yi biz belirliyoruz. 
# bunun için x leri birer polinom olarak tanıtmam lazım fakat bir array e çevirmem lazım 

X_pol = pol.fit_transform(X)

lr = LinearRegression().fit(X_pol,y)
tahmin = lr.predict(X_pol)
r2pol = mt.r2_score(y,tahmin)
mse = mt.mean_squared_error(y,tahmin)
print(f"R2 skoru: {r2pol}\nMSE skoru: {mse}")
# 2. dereceden bu şekilde bir yapıya ulaştı. 
```

```python
plt.scatter(X,y,color="red")
plt.plot(X,tahmin,color="blue")
# çıktı da bu şekilde polinomal bir şekilde oluşmuş oldu. 
# bu şekilde modelimizi en iyi şekilde açıklayan modeli bulmaya çalışıyoruz 
```
