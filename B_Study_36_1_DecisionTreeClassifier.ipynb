```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\anlasekon\Python ile Makine Öğrenmesi\BreastCancerWisconsin.csv")
data2 = data.copy()

data2.head()
```

```python
data2.drop(columns=["id","Unnamed: 32"],axis=1,inplace=True)
data2.diagnosis = [1 if kod=="M" else 0 for kod in data2.diagnosis]
data2.head()
```

```python
y = data2["diagnosis"]
X = data2.drop(columns="diagnosis",axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
preds = model.predict(X_test)
print(f"Decision tree accuracy score: % {accuracy_score(y_test,preds)*100}")
```

```python
names = list(X.columns)
plot_tree(model,filled=True,fontsize=10,feature_names=names,max_depth=1)
plt.show()
```
